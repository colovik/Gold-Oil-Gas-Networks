# Network analysis

```python
import networkx as nx
```

```python
correlation_matrix = ds[['Close*_Gold', 'Close*_Oil', 'Close*_NaturalGas']].corr()
correlation_matrix
```

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
graph = nx.Graph()
```

```python
#Add nodes
variables = ['Gold', 'Oil', 'NaturalGas']
for variable in variables:
    graph.add_node(variable)
```

```python
#Add edges
for i in range(len(variables)):
    for j in range(i + 1, len(variables)):
        correlation = correlation_matrix.iloc[i, j]
        graph.add_edge(variables[i], variables[j], weight=correlation)
```

```python
pos = nx.circular_layout(graph)

nx.draw_networkx_nodes(graph, pos, node_color='red', node_size=2500)

# Add labels to edges
edges = graph.edges(data=True)
edge_labels = {(edge[0], edge[1]): f"{edge[2]['weight']:.2f}" for edge in edges}
nx.draw_networkx_edges(graph, pos)
nx.draw_networkx_edge_labels(graph, pos, edge_labels=edge_labels)

# Add labels to nodeSs
nx.draw_networkx_labels(graph, pos, font_size=10, font_color='black')

plt.axis('off')
plt.show()
```

# Centrality Measures

```python
closeness_centrality = nx.closeness_centrality(graph)
betweenness_centrality = nx.betweenness_centrality(graph)
eigenvector_centrality = nx.eigenvector_centrality(graph)

for node in graph.nodes():
    print(f"Node: {node}")
    print(f"Closeness Centrality: {closeness_centrality[node]}")
    print(f"Betweenness Centrality: {betweenness_centrality[node]}")
    print(f"Eigenvector Centrality: {eigenvector_centrality[node]}\n")
```
